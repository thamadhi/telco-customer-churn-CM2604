```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
from sklearn.preprocessing import MinMaxScaler
```

```python
#Load data
data = 'telco-customer-churn-CM2604/data/Telco-Customer-Churn-unprocessed.csv'
df = pd.read_csv(data)
```

```python
#Drop customer ID
df = df.drop('customerID', axis=1)
```

```python
#monthly charges is in float and total charges is an object.
#As deduced from EDA
df.TotalCharges.values
```

```python
#supply the pd.to_numeric function as a index, find field with empty strings
df[pd.to_numeric(df.TotalCharges,errors='coerce').isnull()]
```

```python
#dropping 11 rows from 7043
df1 = df[df.TotalCharges!=' ']
df1.shape
```

```python
#convert total charges to numeric
df1.loc[:, 'TotalCharges'] = pd.to_numeric(df1.TotalCharges)
```

```python
#identify columns with unique values
def unique_col(df):
  for column in df:
    #identify categorical columns
    if df[column].dtypes == 'object':
      print(f'{column} : {df[column].unique()}')
```

```python
unique_col(df1)
```

```python
#Replace 'No internet/phone service' with 'No'
df1.replace('No internet service','No',inplace=True)
df1.replace('No phone service','No',inplace=True)
```

```python
#Convert Yes/No columns to 1/0
yes_no_columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
          'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']

for col in yes_no_columns:
  df1[col]=df1[col].replace({'Yes':1, 'No':0})
```

```python
for col in df1:
  print(f'{col} : {df1[col].unique()}')
```

```python
#Convert gender to 1/0 (Female=1, Male=0)
df1['gender'] = df1['gender'].replace({'Female': 1, 'Male': 0})
```

```python
#One-hot encode remaining categoricals
df2 = pd.get_dummies(data=df1,columns=['InternetService','Contract','PaymentMethod'],dtype=int)
```

```python
#Scale numeric features (0-1 range)
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']
scaler = MinMaxScaler()

df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])

```
